```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

r = requests.get('https://imsdb.com/all-scripts.html')
soup = BeautifulSoup(r.content, 'html.parser')
# print(soup.prettify())
results = soup.findAll("td", {"valign" : "top"})
```

```python
movies = results[2].findAll("p")
```

```python
movies[0].a.attrs['title']
```

```python
for i in movies:
    print(i)
```

```python
mov_dict = {'title':[], 'URL':[]}
```

```python
for i in movies:
    mov_dict['title'].append(i.a.attrs["title"])
    mov_dict['URL'].append("https://imsdb.com" + i.a.attrs["href"].replace(" ", "%20"))
    
```

```python
mov_dict
```

```python
df = pd.DataFrame()
```

```python
df['Title'] = mov_dict['title']
```

```python
df['URL'] = mov_dict['URL']
```

```python
df.to_csv('moviescript.csv')
```
