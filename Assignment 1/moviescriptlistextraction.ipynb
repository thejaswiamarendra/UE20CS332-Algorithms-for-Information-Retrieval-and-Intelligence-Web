```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

r = requests.get('https://imsdb.com/all-scripts.html')
soup = BeautifulSoup(r.content, 'html.parser')
# print(soup.prettify())
results = soup.findAll("td", {"valign" : "top"})
```

```python
movies = results[2].findAll("p")
```

```python
movies[0].a.attrs['title']
```

```python
for i in movies:
    print(i)
```

```python
mov_dict = {'title':[], 'URL':[]}
```

```python
for i in movies:
    title = i.a.attrs["title"]
    mov_dict['title'].append(title)
    mod_title = title.replace(" Script", "")
    mod_title = mod_title.replace(" ", "-")
    mov_dict['URL'].append("https://imsdb.com/scripts/" + mod_title + ".html")
    
```

```python
mov_dict
```

```python
df = pd.DataFrame()
```

```python
df['Title'] = mov_dict['title']
```

```python
df['URL'] = mov_dict['URL']
```

```python
p = 0
for i in df['URL']:
    if requests.get(i).status_code!=200:
        print(i, requests.get(i).status_code)
        df.drop(p, inplace = True)
    print(str(p*100/len(df))[:4]+"%", end='\r')
    p+=1


#I used this snippet to see if all the URLs are valid. Some aren't so I'm manually dropping those rows
```
